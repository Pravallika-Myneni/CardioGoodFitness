```python
focal_df_list = []
non_focal_df_list = []
```

```python
try:
    for focal_file in focal:
        data_samp = pd.read_csv(focal_file, header=None)
        data_samp.columns = ['X', 'Y']
        focal_df_list.append(data_samp)
except FileNotFoundError:
    print(focal_file)
```

```python
try:
    for non_focal_file in non_focal:
        data_samp = pd.read_csv(non_focal_file, header=None)
        data_samp.columns = ['X', 'Y']
        non_focal_df_list.append(data_samp)
except FileNotFoundError:
    print(non_focal_file)
```

## Applying preprocessing Techniques

```python
from scipy import stats
from scipy.signal import savgol_filter
np.set_printoptions(precision=2)
```

### Data Normalization - Z-score

```python
for focal in focal_df_list:
    focal['X'] = stats.zscore(focal['X'])
    focal['Y'] = stats.zscore(focal['Y'])
    
for non_focal in non_focal_df_list:
    non_focal['X'] = stats.zscore(non_focal['X'])
    non_focal['Y'] = stats.zscore(non_focal['Y'])
```

```python
focal_df_list[0].head()
```

```python
focal_df_list_copy[0].head()
```

### Savitzky-Golay Filtering

```python
for focal in focal_df_list:
    focal['X'] = savgol_filter(focal['X'], 25,3)
    focal['Y'] = savgol_filter(focal['Y'], 25,3)
    
for non_focal in non_focal_df_list:
    non_focal['X'] = savgol_filter(non_focal['X'], 25,3)
    non_focal['Y'] = savgol_filter(non_focal['Y'], 25,3)
```

```python
df_x_focal = pd.DataFrame()

for i in range(3750):
    df_x_focal = df_x_focal.append(focal_df_list[i]['X'], ignore_index= True)
df_x_focal['Class'] = "Focal"

df_x_non_focal = pd.DataFrame()

for i in range(3750):
    df_x_non_focal = df_x_non_focal.append(non_focal_df_list[i]['X'], ignore_index= True)
df_x_non_focal['Class'] = "Non-Focal"

df_x = pd.concat([df_x_focal, df_x_non_focal], ignore_index = True)
df_x.sample(4)
```

```python
df_y_focal = pd.DataFrame()

for i in range(3750):
    df_y_focal = df_y_focal.append(focal_df_list[i]['Y'], ignore_index= True)
df_y_focal['Class'] = "Focal"

df_y_non_focal = pd.DataFrame()

for i in range(3750):
    df_y_non_focal = df_y_non_focal.append(non_focal_df_list[i]['Y'], ignore_index= True)
df_y_non_focal['Class'] = "Non-Focal"

df_y = pd.concat([df_y_focal, df_y_non_focal], ignore_index = True)
df_y.sample(4)
```

```python
df_x.shape, df_y.shape
```

```python
df_enc_x = df_x.copy()
df_enc_x['Class'] = df_enc_x['Class'].astype('category')
df_enc_x['Class'] = df_enc_x['Class'].cat.codes

df_enc_y = df_y.copy()
df_enc_y['Class'] = df_enc_y['Class'].astype('category')
df_enc_y['Class'] = df_enc_y['Class'].cat.codes
```

```python
df_x_1 = df_enc_x.sample(frac=1)
df_y_1 = df_enc_y.sample(frac=1)
```
